**Задание 1.** В чём различие между зависимыми и независимыми выборками?


**Зависимые выборки** - такие, между элементами которой можно установить парную связь. Иначе говоря, можно установить пару (то есть одному случаю из выборки X будет соответствовать один и только один случай из выборки Y и наоборот) для каждого наблюдения в двух выборках (и это основание взаимосвязи является важным для измеряемого на выборках признака).

Например, изучение пациентов до и после прохождения лечения (одна выборка - пациенты до лечения, вторая - после прохождения лечения - изучается динамика показателей). Изучение каких-либо объектов в динамике до-после (повторное изучение свойств).

**Независимые (несвязаные) выборки** - такие выборки, в которые объекты исследования набирались независимо друг от друга и не имеют никакой связи между собой. Процедура исследования и полученные результаты по одной выборке никак не влияют на другую.

**Задание 2.** Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

Определим значение параметрических и непараметрических статистических критериев.

Статистический критерий называется **параметрическим**, если он предполагает конкретное распределение генеральной совокупности (чаще всего, нормальное распределение) и использует параметры этой совокупности, такие как среднее, дисперсию, среднеквадратическое отклонение отклонение и др.

**Непараметрические** статистические критерии - такие, которые не базируется на предположении о типе распределения генеральной совокупности и не использует параметры этой совокупности. 

Следовательно, при нормальном распределении выборки предпочтительнее использовать параметрические критерии, т.к. они будут способны с большей достоверностью отвергать нулевую гипотезу, если она неверна. 
Соотстветственно, если выборка задана не нормальным распределением, следует применять непараметрические критерии, т.к. применение параметрических критериев может привести к ошибочным результатам.
