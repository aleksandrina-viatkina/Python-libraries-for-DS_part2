**Вопрос 1.** Расскажите, как работает регуляризация в решающих деревьях, какие параметры мы штрафуем в данных алгоритмах?

В моделях с решающими деревьями под регуляризацией понимают сложность деревьев, т.е. для снижения переобучения мы подбираем оптимальное значение таких параметров, как - количество листьев в дереве, глубина дерева, количество объектов в листе и пр. (все те параметры, которыми мы формируем идеальное дерево, которое дает максимально точный результат без переобучения/с минимальным переобучением).

Таким образом, штрафуются (ограничиваются) параметры, отвечающие за сложность модели с решающими деревьями.

Также в бустинговых алгоритмах используется параметр - коэффициент регуляризации лямбда (reg_lambda). Обычно он принимает значения от 0 до 1 и обозначает снижение вклада каждой новой базовой предсказывающей модели, добавляемой в общую модель градиентного бустинга на очередной итерации. При обучении модели бустинговых моделей, также важно ограничить(настроить по кросс валидации) гиперпараметры. Лучшим вариантом является подбор этих параметров одновременно.



**Вопроc 2.** По какому принципу рассчитывается "важность признака (feature_importance)" в ансамблях деревьев?

Инициализируется массив feature_importances всех нулей размером n_features .

Проходим по дереву: для каждого внутреннего узла, который разбивается по признаку i , вычисляется уменьшение ошибки этого узла, умноженное на количество выборок, которые были направлены в узел, и добавляется это количество к feature_importances[i] .

Уменьшение ошибки зависит от используемого критерия (например, Джини, энтропия, ...).

Это отношение между количеством выборок (объектов), направляемых в узел принятия решения с участием этого признака в любом из деревьев ансамбля, и общим количеством выборок в обучающем наборе.

Признаки, участвующие в узлах верхнего уровня деревьев решений, имеют тенденцию видеть больше выборок (объектов), следовательно, они, наиболее вероятно, будут иметь большую важность.
